```python
# Check shape before cleaning
print("Before:", df.shape)

# Remove missing rows
df = df.dropna()

# Check shape after cleaning
print("After:", df.shape)

# Verify missing values
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Contract", hue="Churn")
plt.title("Churn by Contract Type")
plt.show()
```

## Key Findings

1. Overall churn rate is 26.58%.

2. Customers on month-to-month contracts exhibit the highest churn rate (42.7%).

3. Fiber Optic customers churn at a significantly higher rate (41.9%) than DSL customers (19.0%).

4. Customers using Electronic Check have the highest churn rate (45.3%).

5. Churned customers pay higher monthly charges on average (74.44 vs 61.31).

6. Churned customers have substantially lower tenure (17.98 months) than retained customers (37.65 months).
